```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn import preprocessing
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
%matplotlib inline

sns.set(style='whitegrid', rc={"grid.linewidth": 0.1})
sns.set_context("paper", rc={"font.size":14,"axes.titlesize":14,"axes.labelsize":14})
```

```python
def read_file():
    with open("./data/digimonbasestats_digivolutionsheet.txt") as inf:
        row = []
        counter = 0
        num = 0
        for line in inf:
            line = line.rstrip()
            if line and counter == 0:
                row = []
                line = line.replace(' ', '')
                name = line.split('(')[0]
                type_split = line[line.find("Type:")+5:line.find('.Ability:')].split('/')
                type_1 = type_split[0]
                if len(type_split) > 1:
                    type_2 = type_split[1]
                else:
                    type_2 = ''
                counter = 1
            elif line:
                row.append(line.split()[1])
            else:
                counter = 0
                num += 1
                row = [num, name, type_1, type_2, np.sum(np.array(row, dtype=int))] + list(map(int, row)) + [-1, 'FALSE']
                yield row
        yield [num, name, type_1, type_2, np.sum(np.array(row, dtype=int))] + list(map(int, row)) + [-1, 'FALSE']
```

```python
df = pd.DataFrame(read_file(), columns=['#', "Name", 'Type 1', 'Type 2', 'Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', "Legendary"])
```

```python
df.to_csv("test.csv", index=False)
```

```python
types = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

X = df[types]
le = preprocessing.LabelEncoder()
le.fit(df['Type 1'])
y = le.transform(df['Type 1']) 
```

```python
def custom_legend(fig, colors, labels, legend_location = "center left", legend_boundary = (1,.5), marker="o"):
    # Create custom legend for colors
    patches = [ plt.plot([],[], marker=marker, ms=10, ls="", mec=None, color=colors[i], 
            label="{:s}".format(labels[i]) )[0]  for i in range(len(labels)) ]
    artist = fig.legend(patches, labels, loc=legend_location, bbox_to_anchor=legend_boundary, fontsize=8)
    return artist
```

```python


pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

fig, ax = plt.subplots()
colors = sns.color_palette("hls", len(le.classes_))
lw = 2

for color, i, target_name in zip(colors, range(len(le.classes_)), y):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.55, lw=lw)
plt.legend(loc='best', shadow=False, scatterpoints=1)
#plt.axis('equal')
plt.title('PCA of Pokemon dataset')
lgd1 = custom_legend(ax, colors, le.classes_, marker='o')
plt.show()
```
