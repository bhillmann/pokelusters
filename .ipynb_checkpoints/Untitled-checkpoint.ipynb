```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
%matplotlib inline

sns.set(style='whitegrid', rc={"grid.linewidth": 0.1})
sns.set_context("paper", rc={"font.size":14,"axes.titlesize":14,"axes.labelsize":14})
```

```python
df = pd.read_csv("data/Pokemon.csv")
```

```python
df = df.loc[df['Legendary'] == False]
```

```python
types = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

X = df[types]
le = preprocessing.LabelEncoder()
le.fit(df['Type 1'])
y = le.transform(df['Type 1']) 
```

```python
def custom_legend(fig, colors, labels, legend_location = "center left", legend_boundary = (1,.5), marker="o"):
    # Create custom legend for colors
    patches = [ plt.plot([],[], marker=marker, ms=10, ls="", mec=None, color=colors[labels[i]], 
            label="{:s}".format(labels[i]) )[0]  for i in range(len(labels)) ]
    artist = fig.legend(patches, labels, loc=legend_location, bbox_to_anchor=legend_boundary, fontsize=8)
    return artist
```

```python
colors = {
    'Bug': '#A8B820',
    'Dark': '#705848',
    'Dragon': '#7038F8',
    'Electric': '#F8D030',
    'Fairy': '#EE99AC',
    'Fighting': '#C03028',
    'Fire': '#F08030',
    'Flying': '#A890F0',
    'Ghost': '#705898',
    'Grass': '#78C850',
    'Ground': '#E0C068',
    'Ice': '#98D8D8',
    'Normal': '#A8A878',
    'Poison': '#A040A0',
    'Psychic': '#F85888',
    'Rock': '#B8A038',
    'Steel': '#B8B8D0',
    'Water': '#6890F0'
}
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

fig, ax = plt.subplots()
lw = 2

for i in range(len(le.classes_)):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=colors[le.inverse_transform([i])[0]], alpha=.55, lw=lw)
plt.legend(loc='best', shadow=False, scatterpoints=1)
#plt.axis('equal')
plt.title('PCA of Pokemon dataset')
lgd1 = custom_legend(ax, colors, le.classes_, marker='o')
plt.show()
```

```python
sns.pairplot(data=df[types + ['Type 1']], hue="Type 1", palette=colors) # S marker size
```
